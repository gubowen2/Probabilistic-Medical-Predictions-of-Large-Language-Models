```python
import numpy as np
from scipy import stats
from collections import defaultdict
```

```python
def get_stdev_only(accuracy):
    print(f"Input accuracy: {np.round(accuracy, 3)}")
    accuracy_by_dataset = [[item[0] for item in accuracy], [item[1] for item in accuracy], [item[2] for item in accuracy], [item[3] for item in accuracy], [item[4] for item in accuracy]]
    average = []
    standard_deviation = []
    for acc in accuracy_by_dataset:
        acc = np.array(acc)
        avg = np.mean(acc)
        stdev = np.std(acc, ddof = 1)
        average.append(avg)
        standard_deviation.append(stdev)
    average_accuracy = np.array([np.mean(np.array(accuracy[0])), np.mean(np.array(accuracy[1])), np.mean(np.array(accuracy[2]))])
    print(f"Average accuracy: {np.round(average_accuracy, 3)}")
    total_average = np.mean(average_accuracy)
    total_stdev = np.std(average_accuracy, ddof = 1)
    print(f"Average accuracy by datasets: {np.round(average, 3)}")
    print(f"Standard deviation by datasets: {np.round(standard_deviation, 3)}")
    print(f"Average average accuracy: {round(total_average, 3)}")
    print(f"Avaergae accuracy standard deviation: {round(total_stdev, 3)}")
```

```python
def get_stdev_and_p_value(internal, external):
    internal = np.array(internal)
    external = np.array(external)
    # Calculate the difference between the two lists
    diff = internal - external

    # Calculate the average
    avg_internal = np.mean(internal)
    avg_external = np.mean(external)
    avg_diff = np.mean(diff)
    
    # Calculate the standard deviation
    std_internal = np.std(internal, ddof = 1)
    std_external = np.std(external, ddof = 1)
    std_diff = np.std(diff, ddof = 1)
     
    # Perform a paired t-test
    t_statistic, p_value = stats.ttest_rel(internal, external, alternative = "greater")

    print("Internal Probability:", internal)
    print("Average of Internal Probability:", round(avg_internal, 3))
    print("Standard Deviation of Internal Probability:", round(std_internal, 3))
    print("External Probability:", external)
    print("Average of External Probability:", round(avg_external, 3))
    print("Standard Deviation of External Probability:", round(std_external, 3))
    print("Differences:", diff)
    print("Average of Differences:", round(avg_diff, 3))
    print("Standard Deviation of Differences:", round(std_diff, 3))
    print("Paired t-test p-value:", round(p_value, 3))
```
